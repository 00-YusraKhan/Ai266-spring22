```python
import pandas as pd
 #for data formingfrom sklearn.metrics import accuracy_score # For Checking Accuracy
from sklearn.model_selection import train_test_split # Splitting Data For Train Test
from sklearn import metrics
from sklearn.naive_bayes import MultinomialNB # For Multinomial Naive Bayes Model
from sklearn.model_selection import cross_val_score # For Cross Validation
import warnings
  # warning ignores small error warnings.filterwarnings('ignore')
```

```python
train = pd.read_csv("train.csv")
 #read train File data test = pd.read_csv('test.csv') #read test File data
```

```python
print(train.shape)
 # Tell how many rows and columnprint(test.shape) # Tell how many rows and column e.g 700000 rows and 32 column
```

```python
del train['id']
 #delete id cos the data we train and train file data both start with 0,1,2 so due to same number of data accuracy will increase. del train['f_27']
 #delete column f_27 because this column had string data which we dont need because our whole data set in intdel test['f_27'] #delete column f_27 because this column had string data which we dont need
```

```python
train.head()
 #Show first 5 column 
```

```python
from sklearn.preprocessing import MinMaxScaler
 # use minmax algorithm to drop targettrainDTar = train.drop(columns=['target'])
 # drop target column from train file scaler = MinMaxScaler()
 #scaler is a variable where we pass min max functionprint(scaler.fit(trainDTar))
 # .transform generate new file from trainDTarnewtrain = scaler.transform(trainDTar)
```

```python
lstCol = list(trainDTar.columns)
```

```python
newtrainDF = pd.DataFrame(newtrain, columns=lstCol)
```

```python
newtrainDF
```

```python
X = newtrainDF
y = train['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.model_selection import GridSearchCV

params = {'alpha': [0.6, 0.1, 0.5, 1, 2.0,3.0],
         }

mnb_clf= GridSearchCV(MultinomialNB(), param_grid=params, n_jobs=6, cv=6, verbose=1)
res = mnb_clf.fit(X,y)

print('Train Accuracy : %.3f'%mnb_clf.best_estimator_.score(X_train, y_train))
print('Test Accuracy : %.3f'%mnb_clf.best_estimator_.score(X_test, y_test))
print('Best Accuracy Through Grid Search : %.3f'%mnb_clf.best_score_)
print('Best Parameters : ',mnb_clf.best_params_)
```

```python
nav_clf = MultinomialNB(alpha =0.1)
nav_scores = cross_val_score(nav_clf, X_train, y_train, cv=6)
print('Naive Bayes Scores: ',nav_scores)
nav_mean = nav_scores.mean()
print('Naive Bayes Mean Score: ',nav_mean)
```
